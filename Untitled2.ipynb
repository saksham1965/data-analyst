```python
maxtotal = df['base_total'].max()
maxtotal_pokemon = df[df['base_total'] == maxtotal]
if len(maxtotal_pokemon) > 1:
  maxlegendary_pokemon = maxtotal_pokemon[maxtotal_pokemon['is_legendary'] == 1]
  if not maxlegendary_pokemon.empty:
    maxtotal_pokemon = maxlegendary_pokemon
  else:
      maxgeneration_pokemon = maxtotal_pokemon[maxtotal_pokemon['generation'] == 3]
      if not maxgeneration_pokemon.empty:
        maxtotal_pokemon = maxgeneration_pokemon
        print(maxtotal_pokemon.to_string(index=False))
```
