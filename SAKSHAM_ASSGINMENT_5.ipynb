```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')

print(iris.head())
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', style='species', palette='Set1')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.boxplot(data=iris, x='species', y='sepal_length', ax=axes[0], palette='Set2')
axes[0].set_title('Sepal Length Distribution by Species')
axes[0].set_ylabel('Sepal Length (cm)')

sns.boxplot(data=iris, x='species', y='sepal_width', ax=axes[1], palette='Set2')
axes[1].set_title('Sepal Width Distribution by Species')
axes[1].set_ylabel('Sepal Width (cm)')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(iris, hue='species', palette='Set2', diag_kind='kde')

# Show the plot
plt.suptitle("Pairplot for Bivariate Analysis of Iris Dataset", y=1.02)
plt.show()
```

```python
correlation = iris['sepal_length'].corr(iris['sepal_width'])
print(f"Correlation between Sepal Length and Sepal Width: {correlation:.2f}")
```

```python
sns.set(style="whitegrid")

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(14, 10))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=feature, data=iris, palette='Set3')
    plt.title(f'{feature.capitalize()} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature.replace('_', ' ').capitalize())

plt.tight_layout()
plt.show()
```

```python
sns.set(style='whitegrid')

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(14, 10))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.kdeplot(data=iris, x=feature, hue='species', fill=True, common_norm=False, palette='Set2', alpha=0.5)
    plt.title(f'Density Plot of {feature.replace("_", " ").capitalize()} by Species')
    plt.xlabel(feature.replace("_", " ").capitalize())
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

sns.set(style='whitegrid')

plt.figure(figsize=(14, 10))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=iris, x=feature, kde=True, color='skyblue', edgecolor='black', stat='density')
    plt.title(f'PDF: {feature.replace("_", " ").capitalize()}')
    plt.xlabel(feature.replace("_", " ").capitalize())
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
