```python
import pandas as pd

df = pd.read_csv('/content/car.csv')
df.head()
```

```python
most_expensive_per_make = df.loc[df.groupby('Make')['MSRP'].idxmax()]
print("1. Most expensive car for each car company:")
print(most_expensive_per_make[['Make', 'Model', 'MSRP']], "\n")
```

```python
if 'convertible' in df.columns:
    convertible = df[df['convertible'] == True]
    most_expensive_convertibles = convertible.loc[convertible.groupby('Make')['MSRP'].idxmax()]
    print(most_expensive_convertible[['Make', 'Model', 'MSRP']])
```

```python
if 'Transmission Type' in df.columns:
    automatic_cars = df[df['Transmission Type'].str.lower().str.contains('automatic')]
    print("3. List of all automatic cars:")
    print(automatic_cars[['Make', 'Model', 'Transmission Type']], "\n")
else:
    print("3. Transmission Type column not found in the dataset.\n")
```

```python
if 'Market Category' in df.columns:
    # Replace empty strings with NaN for consistent handling
    df['Market Category'] = df['Market Category'].str.lower().replace('', pd.NA)
    # Use notna() to filter out rows with NaN in 'Market Category'
    Market_Category = df[df['Market Category'].notna()]
    print("4. Market category of each car company:")
    print(Market_Category[['Make', 'Model', 'MSRP']], "\n")
else:
    print("4. Category column not found — cannot identify Market category.\n")
```

```python
avg_MSRP_per_make = df.groupby('Make')['MSRP'].mean().reset_index()
avg_MSRP_per_make.rename(columns={'MSRP': 'Average_MSRP'}, inplace=True)
print("5. Average MSRP of cars for each car company:")
print(avg_MSRP_per_make)
```
